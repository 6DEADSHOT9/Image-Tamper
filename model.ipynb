```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.applications.resnet50 import ResNet50
```

```python
# Load model weights
ResNet50_model = ResNet50(weights='imagenet', include_top=False, input_shape=(224, 224, 3))
# config = tf.keras.models.load_model('Resnet_model/').get_config()
# model = tf.keras.models.load_model('Resnet_model/tf_model.h5')
```

```python
ResNet50_model.summary()
```

```python
for layer in ResNet50_model.layers:
    # layer.trainable = True
    # print((layer.name),layer.trainable,layer.output_shape)
    # print(layer.name.split('_')[0])
    if layer.name.split('_')[0] == 'conv5':
        layer.trainable = True
    else:
        layer.trainable = False
    print(layer.name, layer.trainable, layer.output_shape)
```

```python
model_top = tf.keras.models.Sequential()
model_top.add(tf.keras.layers.Flatten(input_shape=ResNet50_model.output_shape[1:]))
# model_top.add(tf.keras.layers.Dropout(0.2))
# model_top.add(tf.keras.layers.Dense(512, activation='relu'))
model_top.add(tf.keras.layers.Dense(256, activation='relu'))
model_top.add(tf.keras.layers.Dropout(0.1))
model_top.add(tf.keras.layers.Dense(1, activation='sigmoid'))
model_top.summary()
ResNet50_model.summary()
```

```python
Final_model = tf.keras.models.Sequential([ResNet50_model, model_top])
Final_model.summary()
```

```python
Final_model.compile(optimizer=tf.keras.optimizers.Adam(lr=0.0001), loss='binary_crossentropy', metrics=['accuracy'])
```
